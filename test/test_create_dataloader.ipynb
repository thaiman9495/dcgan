```python
import torch
import torchvision
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path
```

```python
path_parent = Path.cwd()
print(path_parent)
```

```python
data_root_folder = path_parent.parent.joinpath("data/celeba")
img_size = 64
batch_size = 128
```

Apply some transformation to the images of the dataset

```python
list_transformation = [torchvision.transforms.Resize(img_size),
                       torchvision.transforms.CenterCrop(img_size),
                       torchvision.transforms.ToTensor(),
                       torchvision.transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))]
data_transoform = torchvision.transforms.Compose(list_transformation)
```

```python
dataset = torchvision.datasets.ImageFolder(root=data_root_folder, transform=data_transoform)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
batch_example = next(iter(dataloader))
```
